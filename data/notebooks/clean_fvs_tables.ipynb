```python
import pandas as pd
import numpy as np
file_path = '/Users/mihiarc/work/repos/fvs-python/data/'
```

# Crown Ration Equations and Parameters

```python
acr_equations = pd.read_csv(file_path + 'archive/acr_equations.csv')

species_crown_ratio = pd.read_csv(file_path + 'archive/species_crown_ratio.csv')

# # sort by species_code
species_crown_ratio = species_crown_ratio.sort_values(by='species_code')

# merge on species_code and acr_equation_number
species_crown_ratio = species_crown_ratio.merge(acr_equations, on=['species_code', 'acr_equation_number'])

# # write to csv
species_crown_ratio.to_csv(file_path + 'fvs_sn_tables/species_crown_ratio_with_acr.csv', index=False)

species_crown_ratio
```

# Ecounit Codes

```python
base_ecounit_codes = pd.read_csv(file_path + 'archive/base_ecounit_codes.csv')

# write to csv
base_ecounit_codes.to_csv(file_path + 'fvs_sn_tables/base_ecounit_codes.csv', index=False)

#primary key is fvspy_ecounit
base_ecounit_codes
```

```python
ecounit_codes = pd.read_csv(file_path + 'archive/ecounit_codes.csv')

# rename fvspy_ecounit to fvspy_base_ecounit
ecounit_codes = ecounit_codes.rename(columns={'fvspy_ecounit': 'fvspy_base_ecounit'})

# sort by fvs_spcd
ecounit_codes = ecounit_codes.sort_values(by='fvs_spcd')

# drop fvs_ecounit
ecounit_codes = ecounit_codes.drop(columns=['fvs_ecounit'])

# drop last column
ecounit_codes = ecounit_codes.drop(columns=ecounit_codes.columns[-1])

# if 0.000000 replace with 0.0000001
ecounit_codes = ecounit_codes.replace(0.000000, 0.000001)

# rename columns
ecounit_codes = ecounit_codes.rename(columns={'ecounit_b0': 'ecounit_m221', 
                                              'ecounit_b1': 'ecounit_m222',
                                              'ecounit_b2': 'ecounit_t231',
                                              'ecounit_b3': 'ecounit_221',
                                              'ecounit_b4': 'ecounit_222',
                                              'ecounit_b5': 'ecounit_231t',
                                              'ecounit_b6': 'ecounit_231l',
                                              'ecounit_b7': 'ecounit_232',
                                              'ecounit_b8': 'ecounit_234',
                                              'ecounit_b9': 'ecounit_255',
                                              'ecounit_b10': 'ecounit_411'})

# write to csv
ecounit_codes.to_csv(file_path + 'fvs_sn_tables/ecounit_codes.csv', index=False)

ecounit_codes
```

# Small Trees

```python
species_data = pd.read_csv(file_path + 'archive/species_data.csv')

# create df for starts with small_tree
small_tree_growth_parameters = species_data[['species_code'] + [col for col in species_data.columns if col.startswith('small_tree')]]

# sort by species_code
small_tree_growth_parameters = small_tree_growth_parameters.sort_values(by='species_code')

# write to csv
small_tree_growth_parameters.to_csv(file_path + 'fvs_sn_tables/small_tree_growth_parameters.csv', index=False)

small_tree_growth_parameters
```

# Large Trees

```python
species_data = pd.read_csv(file_path + 'archive/species_data.csv')

ln_dds_species_growth_parameters = pd.read_csv(file_path + 'archive/ln_dds_species_growth_parameters.csv')

# define df with species_code, FIA_code, and starts with 'ln_dds'
large_tree_growth_parameters2 = ln_dds_species_growth_parameters[['species_code', 'FIA_code'] + [col for col in ln_dds_species_growth_parameters.columns if col.startswith('ln_dds')]]

# rename FIA_code to fia_spcd
large_tree_growth_parameters2 = large_tree_growth_parameters2.rename(columns={'FIA_code': 'fia_spcd'})

# rename columns that start with ln_dds to start with large_tree
large_tree_growth_parameters2 = large_tree_growth_parameters2.rename(columns=lambda x: x.replace('ln_dds', 'large_tree'))

# rename large_tree_b2=8 to large_tree_b8
large_tree_growth_parameters2 = large_tree_growth_parameters2.rename(columns={'large_tree_b2=8': 'large_tree_b8'})

# drop fia_spcd
large_tree_growth_parameters2 = large_tree_growth_parameters2.drop(columns=['fia_spcd'])

large_tree_growth_parameters2
# create df for starts with ln_dds call it large_tree
large_tree_growth_parameters = species_data[['species_code'] + [col for col in species_data.columns if col.startswith('ln_dds')]]

# rename ln_dds to large_tree
large_tree_growth_parameters1 = large_tree_growth_parameters.rename(columns=lambda x: x.replace('ln_dds', 'large_tree'))

# merge large tree growth parameters1 and large tree growth parameters2 using species_code
large_tree_growth_parameters = large_tree_growth_parameters1.merge(large_tree_growth_parameters2, on='species_code')

# replace 0 with 0.0000001
large_tree_growth_parameters = large_tree_growth_parameters.replace(0, 0.000001)

# sort by species_code
large_tree_growth_parameters = large_tree_growth_parameters.sort_values(by='species_code')

# write to csv
large_tree_growth_parameters.to_csv(file_path + 'fvs_sn_tables/large_tree_growth_parameters.csv', index=False)

# species_data

large_tree_growth_parameters
```
